Import the libraries and data

#### Week 3

lineaire regressie is een verband tussen 2 variabelen waarbij je aan de hand van 1 variabele de andere kan voorspellen. Dit omdat die variabele altijd een lineaire verband heeft met de variabele. Dit betekent dat als de x groter wordt dat die kan stijgen of kan dalen.

```python
df_sub=df.loc[:,"id":"weight"]
df_sub2=df.loc[:,"steps_mean"]
df=pd.concat([df_sub,df_sub2],axis=1)

df["h_m"]=df["height"]/100
df["BMI"]=df["weight"]/(df["h_m"]**2)
df.head()
```

```python
# dummy's maken
df_dum = pd.get_dummies(df,columns=["gender","age","city"])
df_dum.head()
```

```python
df_r=df_dum.dropna()
```

```python
#df_z=df_r.drop(['hh_size', 'id',"education","education_1"], axis=1)
```

```python
df_z=df_r
```

##### LM model

```python
from sklearn.model_selection import train_test_split #this is the function we need to split the data
from sklearn.linear_model import LinearRegression #remember, this is linear regression model from sk-learn
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import math

x = df_z.loc[:, ~df_z.columns.isin(["steps_mean",'hh_size', 'id',"education","education_1"])] #get the weight & height
y = df_z['steps_mean'] #get the steps_mean

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=80) #splits the data into train (70%) and test (30%) data
```

```python
lm = LinearRegression() #this creates a new LR model
lm.fit(x_train, y_train) #this "fits" the model
```

```python
y_test_steps= lm.predict(x_test)
```

```python
r2_score(y_test, y_test_steps)
```

```python
math.sqrt(mean_squared_error(y_test, y_test_steps))
```

De conclusie is dat mijn model 2,4% goed voorspelt. Dat betekent dat de gemiddelde stappen per dag erg lastig te voorspellen is op basis van de gegeven kolommen. De gemiddelde afwijking is ongeveer 3000, dat betekent dat het model 3000 van de voorspelde gemiddelde stappen afzit. Dit laatste is best logisch, de eerste is erg slecht.
